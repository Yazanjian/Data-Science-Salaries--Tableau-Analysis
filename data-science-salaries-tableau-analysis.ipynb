```python
df = pd.read_csv('/kaggle/input/data-science-salaries-2023/ds_salaries.csv')
df.head()
```

```python
fig, ax = plt.subplots()

ax = sns.countplot(ax=ax, data=df, x=df.employment_type, hue=df.work_year)
ax.plot()
```

```python
#Filter based on Year 2023, and grioup by job title 
filtered = df[df['work_year']==2023].groupby('job_title').size().nlargest(10).sort_values(ascending=False).reset_index(name='job_count')
filtered
```

```python
fig, ax = plt.subplots()

ax = sns.barplot(ax=ax, data=filtered, y=filtered.job_title, x=filtered.job_count)

ax.plot()
```
